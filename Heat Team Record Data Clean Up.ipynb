```python
import os
import pandas as pd
import re
```

```python
current_directory = os.getcwd()

files_in_directory = os.listdir(current_directory)

print(files_in_directory)
```

```python
file_name = 'Miami Heat Team Record.csv'
MIA_TR = pd.read_csv(file_name, encoding='ISO-8859-1')
print(MIA_TR)
```

```python
# Read the file line by line and split using the custom separator
data = []
with open(file_name, 'r', encoding='ISO-8859-1') as file:
    for line in file:
        row = re.split(r',\s*', line.strip())
        data.append(row)

# Convert the data to a DataFrame
MIA_TR = pd.DataFrame(data)

# If the first row contains headers, set them as column names
MIA_TR.columns = MIA_TR.iloc[0]
MIA_TR = MIA_TR[1:]

print(MIA_TR)
```

```python
# List all the different columns
columns_list1 = MIA_TR.columns.tolist()

# Print the list of columns
print(columns_list1)
```

```python
MIA_TR['Season'] = MIA_TR['"Season'].str.replace('"', '')
print(MIA_TR)
```

```python
MIA_TR.to_csv('Miami Heat Team Records.csv', index=False)
```
